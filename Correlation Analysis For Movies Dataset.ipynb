```python
#importing necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading in the data
df = pd.read_csv("movies.csv")
```

```python
#checking for the number of columns and rows in the dataset
df.shape
```

```python
#getting a summary of information on numerical columns of the data
df.describe()
```

```python
#checking for the datatypes of the columns
df.dtypes
```

```python
#checking for the null and non null counts of columns
df.info()
```

```python
df['score'].isnull().sum()
```

```python
#checking for the sum of null values in each column
df.isna().sum()
```

```python
#dropping the null values in the dataset
df = df.dropna()
```

```python
#cross-checking whether null values have been dropped
df.isna().sum()
```

```python
#converting the datatype of the votes column to integer
df['votes'] = df['votes'].astype('int')
```

```python
df
```

```python
#checking for the new shape of the data
df.shape
```

```python
#dropping the released column due to inconsistencies
df.drop(['released'], axis=1, inplace=True)
```

```python
df
```

```python
#sorting the data by the gross column in descending order
df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# pd.set_option('display.max_rows', None)
```

```python
#dropping duplicate rows in dataset 
df.drop_duplicates()
```

```python
#projecting a scatterplot of Budget against Gross Earnings
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
#plotting budget vs gross-earnings using seaborn
sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":'red'}, line_kws={"color":"blue"})
```

```python
#checking the correlation of the numerical columns using Pearson Correlation Coefficient
df.corr(method='pearson') #Pearson by default
```

```python
#plotting a correlation matrix as heatmap for numeric features
fig, ax = plt.subplots(figsize=(20,10))
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel("Movie Features")
plt.xlabel("Movie Features")

plt.show()
```

```python
#assigning codes to colums in the datatype category, to help correlation
df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num
```

```python
#plotting a correlation matrix as heatmap for all values in the data.
fig, ax = plt.subplots(figsize=(20, 10))
correlation_matrix = df_num.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel("Movie Features")
plt.xlabel("Movie Features")

plt.show()
```

```python
df_num.corr()
```

```python
#checking for the correlation between different columns
correlation_mat = df_num.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#sorting data for columns with highest correlation
high_corr = sorted_pairs[(sorted_pairs>0.5)]
high_corr
```

We can infer that gross and budget have the highest correlation and therefore it is more likely that the amount of money spent on the production of a movie, positively affects gross earnings.
